```python
import DG_encoding as DG
import torch
# Test
torch.manual_seed(42)
layer = DG.GridCellLayer(16)
x = torch.randn(1, 16, 100, 100)
y = layer(x)

print("Input shape :", x.shape)
print("Output shape:", y.shape)
```

```python
# Create visualization
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2, 4, figsize=(12, 6))
fig.suptitle('Grid Cell Layer Activations (different channels)', fontsize=14)

for i in range(8):
    ax = axs[i//4, i%4]
    im = ax.imshow(y[0, i].detach().cpu().numpy(), cmap='RdBu_r')
    ax.set_title(f'Channel {i}')
    ax.axis('off')
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)

plt.tight_layout()
plt.savefig('grid_cell_activations.png')
print('Plot saved as grid_cell_activations.png')
```
